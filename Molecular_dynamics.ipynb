```python
# --- Clean PDB → SwissParam-ready MOL2 workflow ---
from openbabel import pybel
from google.colab import files
import os

# List of original ligand PDBs
ligands = [
    "Epicatechin_gallate.pdb",
    "Genistein.pdb",
    "Homoeriodictyol.pdb",
    "Sakuranetin.pdb",
    "Stigmasterol.pdb"
]

for pdb_file in ligands:
    mol_name = pdb_file.replace(".pdb","")

    # Load all molecules in PDB
    mols = list(pybel.readfile("pdb", pdb_file))
    print(f"{pdb_file}: {len(mols)} molecules found")

    # Keep only the first molecule/conformer
    mol = mols[0]

    # Add hydrogens
    mol.addh()

    # Generate 3D coordinates
    mol.make3D()

    # Save as clean MOL2
    mol2_file = f"{mol_name}_clean.mol2"
    mol.write("mol2", mol2_file, overwrite=True)

    # Check file size
    size = os.path.getsize(mol2_file)/1024
    print(f"✅ {mol2_file} created, size: {size:.1f} KB")

    # Trigger download
    files.download(mol2_file)
```

```python
from google.colab import files
import os
import glob

# List of original ligand PDBs
ligands = [
    "Epicatechin_gallate.pdb",
    "Genistein.pdb",
    "Homoeriodictyol.pdb",
    "Sakuranetin.pdb",
    "Stigmasterol.pdb"
]

for pdb_file in ligands:
    base_name = pdb_file.replace(".pdb","")

    # Remove old MOL2 files for this ligand
    for f in glob.glob(f"{base_name}*.mol2"):
        os.remove(f)

    # Convert using OpenBabel: split multiple molecules, add hydrogens, 3D
    !obabel {pdb_file} -O {base_name}_clean.mol2 --gen3d -h -m

    # Detect first output file (OpenBabel may append 1,2,...)
    mol2_candidates = sorted(glob.glob(f"{base_name}_clean*.mol2"))
    if not mol2_candidates:
        raise FileNotFoundError(f"No MOL2 generated for {pdb_file}")

    # Take the first one as our clean ligand
    mol2_file = mol2_candidates[0]

    # Rename to standard clean name
    clean_name = f"{base_name}_clean.mol2"
    if mol2_file != clean_name:
        os.rename(mol2_file, clean_name)

    # Delete any extra files
    for f in mol2_candidates[1:]:
        os.remove(f)

    # Check file size
    size = os.path.getsize(clean_name)/1024
    print(f"✅ {clean_name} created, size: {size:.1f} KB")

    # Download
    files.download(clean_name)
```

```python
import pymol2

protein = "2SRC.pdb"
ligands = [
    "Epicatechin_gallate.pdb",
    "Genistein.pdb",
    "Homoeriodictyol.pdb",
    "Sakuranetin.pdb",
    "Stigmasterol.pdb"
]

for lig in ligands:
    out = f"complex_{lig.replace('.pdb','')}.pdb"
    with pymol2.PyMOL() as pymol:
        pymol.cmd.load(protein, "protein")
        pymol.cmd.load(lig, "ligand")
        pymol.cmd.save(out, "protein ligand")
    print(f"✅ Saved {out}")
```

```python
from Bio import PDB

parser = PDB.PDBParser(QUIET=True)
structure = parser.get_structure("protein", "2SRC.pdb")

io = PDB.PDBIO()
class SelectProtein(PDB.Select):
    def accept_residue(self, residue):
        return PDB.is_aa(residue)  # keep only standard amino acids

io.set_structure(structure)
io.save("2SRC_clean.pdb", SelectProtein())
```

```python
from Bio import PDB

parser = PDB.PDBParser(QUIET=True)
structure = parser.get_structure("protein", "2SRC.pdb")

io = PDB.PDBIO()

class SelectStandardProtein(PDB.Select):
    def accept_residue(self, residue):
        # keep only standard 20 amino acids
        return PDB.is_aa(residue, standard=True)

io.set_structure(structure)
io.save("2SRC_clean_standard.pdb", SelectStandardProtein())
print("Saved cleaned PDB with only standard amino acids.")
```
